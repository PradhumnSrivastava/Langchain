### **GROQ API**

```python
#using api used from .env file
from dotenv import load_dotenv
load_dotenv()

#importing necessary libraries and using the API key from .env file to authenticate with the Groq API
import os
#importing the ChatGroq class from the langchain_groq library to interact with the Groq API
from langchain_groq import ChatGroq

#creating an instance of the ChatGroq class with the specified model and invoking it with a prompt to explain AI, then printing the content of the response
llm = ChatGroq(model="llama-3.1-8b-instant", temperature=0.7, max_tokens=100)
#printing the response from the model when asked to explain AI
print(llm.invoke("Explain AI").content)
```

### **From HuggingFace**

```python
#using api used from .env file
from dotenv import load_dotenv
load_dotenv()
#importing necessary libraries and using the API key from .env file to authenticate with the Hugging Face API
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint

#creating an instance of the HuggingFaceEndpoint class with the specified repository ID, temperature, and maximum number of new tokens, then creating an instance of the ChatHuggingFace class with the Hugging Face endpoint and invoking it with a prompt to ask about the capital of France, finally printing the content of the response
llm = HuggingFaceEndpoint(
    repo_id="meta-llama/Meta-Llama-3-8B-Instruct",
    temperature=0.7,
    max_new_tokens=20,
)
#creating an instance of the ChatHuggingFace class with the Hugging Face endpoint and invoking it with a prompt to ask about the capital of France, finally printing the content of the response
model = ChatHuggingFace(llm=llm)
#printing the response from the model when asked about the capital of France
print(model.invoke("What is the capital of France?").content)
```

### **Run Locally From Hugging Face**

```python
from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline

llm = HuggingFacePipeline.from_model_id(
    model_id ="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    task="text-generation",
    pipeline_kwargs=dict(
        max_new_tokens=512,
        do_sample=False,
        repetition_penalty=1.03,
    ),
)   
chat_model = ChatHuggingFace(llm=llm)
response = chat_model.invoke("What is the capital of France?")
print(response.content)
```

## **Messages**

```python
#using api used from .env file
from dotenv import load_dotenv
load_dotenv()

#importing necessary libraries and using the API key from .env file to authenticate with the Hugging Face API, then creating an instance of the HuggingFaceEndpoint class with the specified repository ID, temperature, and maximum number of new tokens, followed by creating an instance of the ChatHuggingFace class with the Hugging Face endpoint and invoking it with a list of messages, finally printing the content of the response
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

#creating an instance of the HuggingFaceEndpoint class with the specified repository ID, temperature, and maximum number of new tokens, then creating an instance of the ChatHuggingFace class with the Hugging Face endpoint and invoking it with a list of messages, finally printing the content of the response
llm = HuggingFaceEndpoint(
    repo_id="meta-llama/Meta-Llama-3-8B-Instruct",
    temperature=0.7,
    max_new_tokens=200,
)

#creating an instance of the ChatHuggingFace class with the Hugging Face endpoint and invoking it with a list of messages, finally printing the content of the response
model = ChatHuggingFace(llm=llm)

#creating a list of messages with a system message and a human message, then invoking the model with the list of messages and printing the content of the response
messages = [
    SystemMessage(content="You are a helpful Philosopher."),
    HumanMessage(content="What is AI?")]

#invoking the model with the list of messages and printing the content of the response
response = model.invoke(messages)
messages.append(AIMessage(content=response.content))

#printing the content of the response from the model when asked about AI
print(response.content)
print(messages)
```

### **Prompt template with messages**

```python
# --------- STEP 1: Load ENV ---------
from dotenv import load_dotenv
load_dotenv()



# --------- STEP 2: Imports ---------
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace
from langchain_core.prompts import ChatPromptTemplate



# --------- STEP 3: Model Setup ---------
llm = HuggingFaceEndpoint(
    repo_id="meta-llama/Meta-Llama-3-8B-Instruct",
    temperature=0.7,
    max_new_tokens=200,
)
model = ChatHuggingFace(llm=llm)


# --------- STEP 4: Template + Messages ---------
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful AI teacher."),
    ("human", "Explain {topic} in {style}")
])

# --------- STEP 5: Input Values ---------
topic = "Artificial Intelligence"
style = "simple words"

# --------- STEP 6: Create Messages ---------
messages = chat_template.format_messages(
    topic=topic,
    style=style
)

# --------- STEP 7: Call Model ---------
response = model.invoke(messages)

# --------- STEP 8: Output ---------
print(response.content)
```

### **String Output Parser**

```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

load_dotenv()

llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0.9,
    max_tokens=2000
)

parser = StrOutputParser()

# First step
template1 = PromptTemplate(
    input_variables=["input"],
    template="Explain {input} in detail."
)

# Second step
template2 = PromptTemplate(
    input_variables=["text"],
    template="Summarize {text} in 3 lines."
)

# Correct chaining
chain = (
    template1 | llm | parser | template2 | llm | parser
)

response = chain.invoke({"input": "black holes"})
print(response)
```
