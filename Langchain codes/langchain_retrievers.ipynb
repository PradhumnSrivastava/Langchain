## **2. Setup Embeddings and Vector Store**

```python
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS

texts = [
    'Machine learning is a subset of artificial intelligence',
    'Deep learning uses neural networks',
    'Data science analyzes data to extract insights',
    'AI powers chatbots and recommendation systems'
]

embeddings = OpenAIEmbeddings()

vectorstore = FAISS.from_texts(texts, embeddings)
```

## **3. Basic Retriever**

```python
retriever = vectorstore.as_retriever()

docs = retriever.invoke('What is machine learning?')

docs
```

## **4. Similarity Search Retriever**

```python
retriever = vectorstore.as_retriever(search_type='similarity', search_kwargs={'k':2})

docs = retriever.invoke('Explain AI')

docs
```

## **5. Max Marginal Relevance (MMR) Retriever**

```python
retriever = vectorstore.as_retriever(search_type='mmr', search_kwargs={'k':2})

docs = retriever.invoke('Explain neural networks')

docs
```

## **6. MultiQueryRetriever**

```python
from langchain.retrievers.multi_query import MultiQueryRetriever
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model='gpt-4o-mini')

retriever = MultiQueryRetriever.from_llm(
    retriever=vectorstore.as_retriever(),
    llm=llm
)

docs = retriever.invoke('Explain machine learning')

docs
```

## **7. Contextual Compression Retriever**

```python
from langchain.retrievers import ContextualCompressionRetriever
from langchain.retrievers.document_compressors import LLMChainExtractor

compressor = LLMChainExtractor.from_llm(llm)

compression_retriever = ContextualCompressionRetriever(
    base_compressor=compressor,
    base_retriever=vectorstore.as_retriever()
)

docs = compression_retriever.invoke('Explain artificial intelligence')

docs
```

## **8. Parent Document Retriever**

```python
from langchain.retrievers import ParentDocumentRetriever
from langchain.storage import InMemoryStore
from langchain_text_splitters import RecursiveCharacterTextSplitter

store = InMemoryStore()

child_splitter = RecursiveCharacterTextSplitter(chunk_size=200)

retriever = ParentDocumentRetriever(
    vectorstore=vectorstore,
    docstore=store,
    child_splitter=child_splitter
)

print('Parent document retriever initialized')
```

## **10. Retriever + RAG Chain**

```python
from langchain.chains import RetrievalQA

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vectorstore.as_retriever()
)

qa_chain.invoke('What is deep learning?')
```
