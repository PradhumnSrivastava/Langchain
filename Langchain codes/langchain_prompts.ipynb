## **3. Basic PromptTemplate**

```python
from langchain.prompts import PromptTemplate

template = "Explain {topic} in simple terms"

prompt = PromptTemplate(
    input_variables=['topic'],
    template=template
)

print(prompt.format(topic='Machine Learning'))
```

## **4. PromptTemplate with LLM**

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model='gpt-4o-mini')

formatted_prompt = prompt.format(topic='Neural Networks')

response = llm.invoke(formatted_prompt)

print(response.content)
```

## **5. ChatPromptTemplate (System + Human Messages)**

```python
from langchain.prompts import ChatPromptTemplate

chat_prompt = ChatPromptTemplate.from_messages([
    ('system','You are an AI teacher'),
    ('human','Explain {topic}')
])

final_prompt = chat_prompt.format_messages(topic='Deep Learning')

print(final_prompt)
```

## **6. Few-Shot Prompting**

```python
from langchain.prompts import FewShotPromptTemplate

examples = [
    {'question':'2+2','answer':'4'},
    {'question':'3+3','answer':'6'}
]

example_prompt = PromptTemplate(
    input_variables=['question','answer'],
    template='Question: {question}\nAnswer: {answer}'
)

few_shot_prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_prompt,
    prefix='Solve the following problems:',
    suffix='Question: {input}',
    input_variables=['input']
)

print(few_shot_prompt.format(input='5+5'))
```

## **7. Dynamic Prompt with Multiple Variables**

```python
template = "Explain {topic} in {level} level"

prompt = PromptTemplate(
    input_variables=['topic','level'],
    template=template
)

print(prompt.format(topic='Artificial Intelligence', level='beginner'))
```

## **8. Prompt + LCEL Chain**

```python
from langchain_core.output_parsers import StrOutputParser

chain = prompt | llm | StrOutputParser()

chain.invoke({'topic':'Data Science','level':'simple'})
```

## **9. Prompt with Output Parser**

```python
from langchain_core.output_parsers import JsonOutputParser
from pydantic import BaseModel

class Person(BaseModel):
    name: str
    age: int

parser = JsonOutputParser(pydantic_object=Person)

prompt = "Generate a random person with name and age in JSON format"

response = llm.invoke(prompt)

print(parser.parse(response.content))
```

## **10. Prompt Reusability in Chains**

```python
explain_prompt = PromptTemplate(
    input_variables=['concept'],
    template='Explain {concept} with example'
)

chain = explain_prompt | llm | StrOutputParser()

chain.invoke({'concept':'Neural Networks'})
```
