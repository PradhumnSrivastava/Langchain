## **2. Basic LLM Setup**

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model='gpt-4o-mini', temperature=0)

llm.invoke('Explain artificial intelligence').content
```

## **3. RunnableLambda (Custom Function)**

```python
from langchain_core.runnables import RunnableLambda

def add_exclamation(text):
    return text + ' !!!'

runnable = RunnableLambda(add_exclamation)

runnable.invoke('Hello world')
```

## **4. RunnablePassthrough (Pass Input Forward)**

```python
from langchain_core.runnables import RunnablePassthrough

runnable = RunnablePassthrough()

runnable.invoke('Test input')
```

## **5. RunnableSequence (Pipeline)**

```python
from langchain_core.output_parsers import StrOutputParser
from langchain.prompts import PromptTemplate

prompt = PromptTemplate(
    template='Explain {topic} simply',
    input_variables=['topic']
)

chain = prompt | llm | StrOutputParser()

chain.invoke({'topic':'Machine Learning'})
```

## **6. RunnableParallel (Run Tasks in Parallel)**

```python
from langchain_core.runnables import RunnableParallel

parallel = RunnableParallel(
    joke=llm,
    poem=llm
)

parallel.invoke('Write something about AI')
```

## **7. RunnableMap (Map Multiple Inputs)**

```python
from langchain_core.runnables import RunnableMap

runnable_map = RunnableMap({
    'explanation': llm,
    'summary': llm
})

runnable_map.invoke('Explain neural networks')
```

## **8. RunnableBranch (Conditional Routing)**

```python
from langchain_core.runnables import RunnableBranch

branch = RunnableBranch(
    (lambda x: 'math' in x.lower(), RunnableLambda(lambda x: 'Math question detected')),
    RunnableLambda(lambda x: 'General question')
)

branch.invoke('Solve math problem')
```

## **9. Streaming with Runnable Pipeline**

```python
chain = prompt | llm

for chunk in chain.stream({'topic':'Deep Learning'}):
    print(chunk.content, end='', flush=True)
```

## **10. Runnable with Dictionary Output**

```python
from langchain_core.runnables import RunnablePassthrough

chain = {
    'topic': RunnablePassthrough(),
    'explanation': prompt | llm
}

chain.invoke('Artificial Intelligence')
```
