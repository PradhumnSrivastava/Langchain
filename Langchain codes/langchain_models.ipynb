## **2. Load API Key (.env)**

```python
import os
from dotenv import load_dotenv

load_dotenv()

OPENAI_API_KEY = os.getenv('OPENAI_API_KEY')
print('API Key Loaded:', OPENAI_API_KEY is not None)
```

## **3. ChatOpenAI Model (OpenAI LLM)**

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model='gpt-4o-mini')

response = llm.invoke('Explain Artificial Intelligence in simple words')
print(response.content)
```

## **4. HuggingFace Model**

```python
from langchain_community.llms import HuggingFaceHub

llm = HuggingFaceHub(
    repo_id='google/flan-t5-large',
    model_kwargs={'temperature':0.5, 'max_length':100}
)

print(llm.invoke('Explain machine learning'))
```

## **5. Local LLM using Ollama**

```python
from langchain_community.llms import Ollama

llm = Ollama(model='llama3')

print(llm.invoke('What is deep learning?'))
```

## **6. Using LLM with PromptTemplate**

```python
from langchain.prompts import PromptTemplate

template = 'Explain {topic} for beginners'

prompt = PromptTemplate(
    input_variables=['topic'],
    template=template
)

final_prompt = prompt.format(topic='Neural Networks')

print(llm.invoke(final_prompt))
```

## **7. Chat Model with Messages**

```python
from langchain_core.messages import HumanMessage

response = llm.invoke([
    HumanMessage(content='What is data science?')
])

print(response)
```

## **8. Streaming Responses (Optional)**

```python
llm = ChatOpenAI(model='gpt-4o-mini', streaming=True)

for chunk in llm.stream('Explain neural networks'):
    print(chunk.content, end='', flush=True)
```

## **9. Temperature Control**

```python
llm = ChatOpenAI(
    model='gpt-4o-mini',
    temperature=0.7
)

print(llm.invoke('Generate a creative AI startup idea').content)
```
