## **2. Load Embeddings Model**

```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings()

embeddings.embed_query('Artificial Intelligence')[:5]
```

## **3. Create FAISS Vector Store**

```python
from langchain_community.vectorstores import FAISS

texts = [
    'Machine learning is a subset of AI',
    'Deep learning uses neural networks',
    'Data science analyzes data'
]

vectorstore = FAISS.from_texts(texts, embeddings)

vectorstore
```

## **4. Similarity Search in FAISS**

```python
results = vectorstore.similarity_search('What is machine learning?')

results
```

## **5. Add Documents to Vector Store**

```python
new_texts = [
    'Artificial Intelligence powers chatbots',
    'Neural networks learn patterns from data'
]

vectorstore.add_texts(new_texts)

vectorstore.similarity_search('chatbots')
```

## **6. Create Chroma Vector Store**

```python
from langchain_community.vectorstores import Chroma

chroma_db = Chroma.from_texts(
    texts,
    embeddings
)

chroma_db
```

## **7. Similarity Search in Chroma**

```python
results = chroma_db.similarity_search('Explain neural networks')

results
```

## **8. Persist Chroma Vector Database**

```python
persistent_db = Chroma.from_texts(
    texts,
    embeddings,
    persist_directory='chroma_db'
)

persistent_db.persist()
```

## **9. Create Retriever from Vector Store**

```python
retriever = vectorstore.as_retriever()

retriever.invoke('Explain AI')
```

## **10. Vector Store + RAG Example**

```python
from langchain.chains import RetrievalQA
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model='gpt-4o-mini')

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever
)

qa_chain.invoke('What is deep learning?')
```
