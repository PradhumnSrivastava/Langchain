## **2. Setup LLM**

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model='gpt-4o-mini', temperature=0)

llm.invoke('Explain AI').content
```

## **3. Create Tool using @tool Decorator**

```python
from langchain.tools import tool

@tool
def multiply(a: int, b: int) -> int:
    '''Multiply two numbers'''
    return a * b

multiply.invoke({'a':5,'b':6})
```

## **4. Tool with Multiple Parameters**

```python
@tool
def calculate_area(length: int, width: int) -> int:
    '''Calculate area of rectangle'''
    return length * width

calculate_area.invoke({'length':10,'width':5})
```

## **5. StructuredTool Example**

```python
from langchain.tools import StructuredTool

def add_numbers(a: int, b: int) -> int:
    return a + b

add_tool = StructuredTool.from_function(add_numbers)

add_tool.invoke({'a':3,'b':7})
```

## **6. Tool with Pydantic Schema**

```python
from pydantic import BaseModel

class MultiplyInput(BaseModel):
    a: int
    b: int

def multiply_numbers(a:int,b:int):
    return a*b

multiply_tool = StructuredTool.from_function(
    func=multiply_numbers,
    args_schema=MultiplyInput
)

multiply_tool.invoke({'a':4,'b':8})
```

## **7. Tool Calling with LLM**

```python
tools = [multiply]

response = llm.invoke(
    'What is 5 multiplied by 9?',
    tools=tools
)

response
```

## **8. Multiple Tools Example**

```python
tools = [multiply, calculate_area]

response = llm.invoke(
    'Calculate area of rectangle with length 8 and width 3',
    tools=tools
)

response
```

## **9. Tool Integration with Agent**

```python
from langchain.agents import initialize_agent
from langchain.agents import AgentType

agent = initialize_agent(
    tools=[multiply, calculate_area],
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

agent.run('What is 7 times 6?')
```

## **10. Tool for External API Example**

```python
@tool
def get_weather(city: str) -> str:
    '''Get weather for a city'''
    return f'Weather in {city} is sunny'

get_weather.invoke({'city':'Delhi'})
```
