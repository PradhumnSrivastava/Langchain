## **2. Basic LLM Setup**

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model='gpt-4o-mini', temperature=0)

response = llm.invoke('Explain machine learning')
print(response.content)
```

## **3. LLMChain (Basic Chain)**

```python
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate

prompt = PromptTemplate(
    template='Explain {topic} in simple words',
    input_variables=['topic']
)

chain = LLMChain(llm=llm, prompt=prompt)

chain.run('Artificial Intelligence')
```

## **4. Simple Sequential Chain**

```python
from langchain.chains import SimpleSequentialChain

prompt1 = PromptTemplate(
    template='Write a short topic about {topic}',
    input_variables=['topic']
)

prompt2 = PromptTemplate(
    template='Explain this topic briefly: {text}',
    input_variables=['text']
)

chain1 = LLMChain(llm=llm, prompt=prompt1)
chain2 = LLMChain(llm=llm, prompt=prompt2)

overall_chain = SimpleSequentialChain(chains=[chain1, chain2])

overall_chain.run('Data Science')
```

## **5. Sequential Chain (Multiple Inputs / Outputs)**

```python
from langchain.chains import SequentialChain

prompt1 = PromptTemplate(
    template='Generate a company name for {product}',
    input_variables=['product']
)

chain1 = LLMChain(llm=llm, prompt=prompt1, output_key='company')

prompt2 = PromptTemplate(
    template='Write a slogan for {company}',
    input_variables=['company']
)

chain2 = LLMChain(llm=llm, prompt=prompt2, output_key='slogan')

overall_chain = SequentialChain(
    chains=[chain1, chain2],
    input_variables=['product'],
    output_variables=['company','slogan']
)

overall_chain({'product':'AI assistant'})
```

## **6. LCEL Chain (Modern LangChain Pipeline)**

```python
from langchain_core.output_parsers import StrOutputParser

prompt = PromptTemplate(
    template='Explain {topic} simply',
    input_variables=['topic']
)

chain = prompt | llm | StrOutputParser()

chain.invoke({'topic':'Deep Learning'})
```

## **7. RetrievalQA Chain (RAG)**

```python
from langchain.chains import RetrievalQA
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings

documents = ['Machine learning is a subset of AI']

embeddings = OpenAIEmbeddings()

vectorstore = FAISS.from_texts(documents, embeddings)

retriever = vectorstore.as_retriever()

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever
)

qa_chain.invoke('What is machine learning?')
```

## **9. Map Reduce Chain (Document Summarization)**

```python
from langchain.chains.summarize import load_summarize_chain

docs = ['LangChain helps build LLM apps', 'It connects models with tools']

chain = load_summarize_chain(llm, chain_type='map_reduce')

chain.run(docs)
```

## **10. Prompt + Output Parser Chain**

```python
from langchain_core.output_parsers import JsonOutputParser

parser = JsonOutputParser()

prompt = PromptTemplate(
    template='Generate JSON about {topic}',
    input_variables=['topic']
)

chain = prompt | llm | parser

chain.invoke({'topic':'Artificial Intelligence'})
```
