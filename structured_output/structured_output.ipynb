## **Structured Output Using Typed Dict**

```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from typing import TypedDict

load_dotenv()

#model
llm = ChatGroq(
    model="llama-3.3-70b-versatile",  #working model
    temperature=0.7,
    max_tokens=100
)

#Schema
class Review(TypedDict):
    name: str
    summary: str
    sentiment: str
    

#Structured output
structured_model = llm.with_structured_output(Review)

#Input
text = "Hare krsna Mahamantra is too good and very beautiful for the heart and make heart pure"

try:
    ans = structured_model.invoke(
        f"Analyze the sentiment and give a short summary:\n{text}"
    )

    print("Summary:", ans["summary"])
    print("Sentiment:", ans["sentiment"])
    print("Sentiment:", ans["name"])
    

except Exception as e:
    print(" Error:", str(e))
```

## **With Annotation**

```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from typing import TypedDict, Annotated

load_dotenv()

#model
llm = ChatGroq(
    model="llama-3.3-70b-versatile",  #working model
    temperature=0.7,
    max_tokens=100
)

#Schema
class Review(TypedDict):
    name: Annotated[str, "Generate a meaningful title starting with 'The'"]
    summary: Annotated[str, "Summarize in maximum 1000 words"]
    sentiment: Annotated[str, "Describe sentiment in 3-5 words"]
    

#Structured output
structured_model = llm.with_structured_output(Review)

#Input
text = "Hare krsna Mahamantra is too good and very beautiful for the heart and make heart pure"

try:
    ans = structured_model.invoke(
        f"Analyze the sentiment and give a short summary:\n{text}"
    )

    print("Summary:", ans["summary"])
    print("Sentiment:", ans["sentiment"])
    print("Sentiment:", ans["name"])
    

except Exception as e:
    print(" Error:", str(e))
```
